```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path().resolve().parents[1]
```

```python
pharma_data = pd.read_csv(BASE_DIR /"data/raw/df_pharma.csv", sep = ",")
```

```python
pharma_data["cp"] = pd.to_numeric(pharma_data["cp"], errors="coerce")
pharma_data["lat"] = pd.to_numeric(pharma_data["lat"], errors="coerce")
pharma_data["lon"] = pd.to_numeric(pharma_data["lon"], errors="coerce")

pharma_data["region"] = pharma_data["region"].str.strip()
```

```python
pharma = pharma_data[["region"]].copy()
pharma["region"] = pharma["region"].astype(str).str.strip()

pharma = pharma[pharma["region"] != "Corse"]
```

```python
pharma_count = pharma.groupby("region").size().reset_index(name="nb_pharmacies")
```

```python
region_name_to_code = {
    "Île-de-France": 11,
    "Centre-Val de Loire": 24,
    "Bourgogne-Franche-Comté": 27,
    "Normandie": 28,
    "Hauts-de-France": 32,
    "Grand Est": 44,
    "Pays de la Loire": 52,
    "Bretagne": 53,
    "Nouvelle-Aquitaine": 75,
    "Occitanie": 76,
    "Auvergne-Rhône-Alpes": 84,
    "Provence-Alpes-Côte d'Azur": 93,
    "Corse": 94}
```

```python
pharma_count["region_code"] = pharma_count["region"].map(region_name_to_code)
```

```python
pharma_count.to_csv(BASE_DIR /"data/processed/pharma_region.csv",sep=";",index=False,encoding="utf-8")
```

```python
df_dens = pd.read_csv(BASE_DIR /"data/processed/df_dens_med.csv", sep=";")
```

```python
df_pharma = pd.read_csv(BASE_DIR /"data/processed/pharma_region.csv", sep=";")

df_pharma = df_pharma.merge(
    df_dens[["region", "population"]],
    left_on="region_code",
    right_on="region",
    how="left")

df_pharma["pharma_pour_100k"] = (df_pharma["nb_pharmacies"] / df_pharma["population"] * 100000).round(0)
```

```python
df_pharma.to_csv(BASE_DIR /"data/processed/pharma_density.csv", sep=";", index=False)
```
