```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path().resolve().parents[1]
```

```python
df_atc1 = pd.read_csv(BASE_DIR / "data/processed/atc1_prescriptions_2014_2024.csv", sep = ";")
df_atc2 = pd.read_csv(BASE_DIR / "data/processed/atc2_prescriptions_2014_2024.csv", sep = ";")
df_pop = pd.read_csv(BASE_DIR / "data/processed/pop_reg_annee.csv", sep = ";")
```

```python
df_atc1 = df_atc1.merge(df_pop, on = ["annee","region"], how = "left")
df_atc2 = df_atc2.merge(df_pop, on = ["annee","region"], how = "left")
```

```python
df_atc1["boites_pour_1000"] = (df_atc1["boites"] / df_atc1["population"] * 1000).round(0).astype(int)
df_atc2["boites_pour_1000"] = (df_atc2["boites"] / df_atc2["population"] * 1000).round(0).astype(int)
```

```python
df_atc1.to_csv(BASE_DIR / "data/processed/atc1_prescriptions_2014_2024.csv", sep = ";", index = False)
df_atc2.to_csv(BASE_DIR / "data/processed/atc2_prescriptions_2014_2024.csv", sep = ";", index = False)
```

```python
df_atc1.to_parquet(BASE_DIR / "data/processed/atc1_prescriptions_2014_2024.parquet", index = False)
df_atc2.to_parquet(BASE_DIR / "data/processed/atc2_prescriptions_2014_2024.parquet", index = False)
```
