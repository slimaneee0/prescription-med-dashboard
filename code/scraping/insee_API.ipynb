```python
import requests
import pandas as pd 
from pathlib import Path

BASE_DIR = Path().resolve().parents[1]
```

```python
url = "https://api.insee.fr/melodi/data/DS_ESTIMATION_POPULATION?UNIT_MEASURE=PS&SEX=_T&AGE=_T&TIME_PERIOD=2024&TIME_PERIOD=2023&TIME_PERIOD=2022&TIME_PERIOD=2021&TIME_PERIOD=2020&TIME_PERIOD=2019&TIME_PERIOD=2018&TIME_PERIOD=2017&TIME_PERIOD=2016&TIME_PERIOD=2015&TIME_PERIOD=2014&GEO=REG"

response = requests.get(url)

data_pop = response.json()
```

```python
data_list = []

for i in data_pop["observations"]:
    data_individual = {"annee" : "", "region" : "", "population" : ""}
    data_individual["region"] = int((i["dimensions"]["GEO"]).split("-")[2])
    data_individual["annee"] = int((i["dimensions"]["TIME_PERIOD"]))
    data_individual["population"] = int((i["measures"]["OBS_VALUE_NIVEAU"]["value"]))

    data_list.append(data_individual)
```

```python
df_pop = pd.DataFrame(data_list)
df_pop = df_pop[~df_pop["region"].isin([94])]
df_pop = df_pop.sort_values(by = ["annee", "region"], ascending= [False, True])
```

```python
df_pop.to_csv(BASE_DIR / "data/processed/pop_reg_annee.csv", sep = ";", index = False)
df_pop.to_parquet(BASE_DIR / "data/processed/pop_reg_annee.parquet", index = False)
```
